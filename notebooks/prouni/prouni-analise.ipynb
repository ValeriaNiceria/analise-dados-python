```python
# importando a biblioteca pandas
import pandas as pd
```

```python
# Carregando o arquivo CSV
df = pd.read_csv('cursos-prouni.csv')
```

```python
# Análise matemática das colunas numéricas
df.describe()
```

```python
# Descrever os dados
df.info()
```

```python
# Quantidade de cursos por UF
df['uf_busca'].value_counts()
```

```python
# Qual o curso com a mensalidade mais alta
df[df['mensalidade'] == df['mensalidade'].max()]
```

```python
# Qual o curso com a menor mensalidade
df[df['mensalidade'] == df['mensalidade'].min()][['uf_busca', 'universidade_nome', 'mensalidade']]
```

```python
# Qual mensalidade mais alta por UF
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# Mostrar todos os dados dos cursos mais caros de cada UF
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()][['uf_busca', 'universidade_nome', 'nome', 'mensalidade']]
```

```python
# Mensalidades entre 500 e 1000
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)][['uf_busca', 'universidade_nome', 'nome', 'mensalidade']]
```

```python
# Quais as 5 maiores mensalidades
df.nlargest(5, 'mensalidade')[['uf_busca', 'universidade_nome', 'nome', 'mensalidade']]
```

```python
# Quais as 5 menores mensalidades
df.nsmallest(5, 'mensalidade')[['uf_busca', 'universidade_nome', 'nome', 'mensalidade']]
```
